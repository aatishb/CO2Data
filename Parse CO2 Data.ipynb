```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Mauna Loa Observatory Data (Keeling Curve)
### Used for 1959 - 2019
Data accessed from https://library.ucsd.edu/dc/object/bb3859642r

```python
df = pd.read_csv ('Mauna Loa Observatory (present day - 1959).csv', na_values=-99.99, skipinitialspace = True)
annualMean = df.groupby('Yr').mean()
annualMean
```

```python
ax = plt.gca()

for yax in ['CO2 (ppm)', 'CO2 (ppm) seasonally adjusted', 'CO2 (ppm) fit', 'CO2 (ppm)  seasonally adjusted fit','CO2 (ppm) filled', 'CO2 (ppm) seasonally adjusted filled']:
    annualMean.plot(y = yax, figsize=(15,10), ax=ax)

plt.show()
```

```python
exportData1 = annualMean[['CO2 (ppm)']][1:-1].rename(columns = {'CO2 (ppm)':'CO2'})
exportData1.index.names = ['YearAD']
exportData1
```

# Law Dome Ice Core Data
### Used for 1 - 1958
Data accessed from https://www.ncdc.noaa.gov/paleo-search/study/9959

```python
df = pd.read_csv('Law Dome Ice Core 2000 Year (1 - 2004).txt', delim_whitespace=True)
df.plot(x = 'YearAD', y='CO2spl',figsize=(15,10))
plt.show()
```

```python
exportData2 = df[['YearAD', 'CO2spl']].set_index('YearAD')[0:-46].rename(columns = {'CO2spl':'CO2'})
exportData2
```

```python
ax = plt.gca()

exportData1.plot(figsize=(15,10), ax=ax)
exportData2.plot(figsize=(15,10), ax=ax)

plt.show()
```

# Ice Core 8000KYr
### Used for past 800 thousand years until year -1
Data accessed from https://www.ncdc.noaa.gov/paleo-search/study/6091

```python
df = pd.read_csv ('Ice Core 8000KYr (1813 - 800KYr).txt', delim_whitespace=True)
df['YearAD'] = 1950 - df['Age(yrBP)']
exportData3 = df.set_index('YearAD')[['CO2(ppmv)']][21:].rename(columns = {'CO2(ppmv)':'CO2'})
exportData3
```

```python
ax = plt.gca()

exportData1.plot(figsize=(15,10), ax=ax)
exportData2.plot(figsize=(15,10), ax=ax)
exportData3.plot(figsize=(15,10), ax=ax)

ax.set_xlim(-800000,2000)
plt.show()
```

```python
exportData = pd.concat([exportData1, exportData2, exportData3]).sort_index(ascending = False)
exportData
```

```python
exportData.plot(figsize=(15,10), xlim=[2019,-800000])
plt.show()
```
