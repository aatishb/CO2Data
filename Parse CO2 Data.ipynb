```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Mauna Loa Observatory Data (Keeling Curve)
### Used for 1959 - 2019 CE
Data accessed from https://library.ucsd.edu/dc/object/bb3859642r

```python
df = pd.read_csv ('Mauna Loa Observatory 60 Year CO2 Data.csv', na_values=-99.99, skipinitialspace = True)
keeling = df.groupby('Yr').mean()
keeling
```

### Graph showing Keeling Data

```python
ax = plt.gca()

for yax in ['CO2 (ppm)', 'CO2 (ppm) seasonally adjusted', 'CO2 (ppm) fit', 'CO2 (ppm)  seasonally adjusted fit','CO2 (ppm) filled', 'CO2 (ppm) seasonally adjusted filled']:
    keeling.plot(y = yax, figsize=(15,10), ax=ax)

plt.title('Mauna Loa Observatory Data (Keeling Curve)')
plt.show()
```

### Preparing Keeling Data for Export

```python
exportData1 = keeling[['CO2 (ppm)']][1:-1].rename(columns = {'CO2 (ppm)':'CO2'})
exportData1.index.names = ['YearAD']
exportData1
```

# Law Dome Ice Core Data
### Used for 1 - 1958 CE
Data accessed from https://www.ncdc.noaa.gov/paleo-search/study/9959

### Graph Showing Law Dome Data

```python
lawdome = pd.read_csv('Law Dome Ice Core 2000 Year CO2 Data.txt', delim_whitespace=True)
lawdome.plot(x = 'YearAD', y='CO2spl',figsize=(15,10))
plt.title('Law Dome Ice Core Data')
plt.show()
```

### Graph Showing Overlap of Law Dome and Keeling Data 

```python
ax = plt.gca()
lawdome.plot(x = 'YearAD', y='CO2spl',figsize=(15,10), ax=ax, label = 'Law Dome')
keeling.plot(y = 'CO2 (ppm)', figsize=(15,10), ax=ax, label = 'Keeling')

plt.title('Law Dome Ice Core & Keeling Curve Data')
plt.show()
```

### Preparing Law Dome Data for Export

```python
exportData2 = lawdome[['YearAD', 'CO2spl']].set_index('YearAD')[0:-46].rename(columns = {'CO2spl':'CO2'})
exportData2
```

### Graph showing exported portions of Law Dome and Keeling Data 

```python
ax = plt.gca()

exportData1.plot(figsize=(15,10), ax=ax, label = 'Keeling')
exportData2.plot(figsize=(15,10), ax=ax, label = 'Law Dome')

plt.title('Selected Portions of Law Dome and Keeling Data')
plt.show()
```

# EPICA Dome C Ice Core Data
### Used for past 800 thousand years until Year 0
Data accessed from https://www.ncdc.noaa.gov/paleo-search/study/6091

Note that BP = Before Present = Before 1950. So before exporting the data we need to convert the year to be consistent with the other data sets

### Preparing Data for Export

```python
domec = pd.read_csv ('Dome C Ice Core 800KYr CO2 Data.txt', delim_whitespace=True)
domec['YearAD'] = 1950 - domec['Age(yrBP)']
exportData3 = domec.set_index('YearAD')[['CO2(ppmv)']][21:].rename(columns = {'CO2(ppmv)':'CO2'})
exportData3
```

### Graphing all Data Together (Keeling, Law Dome, Dome C)

```python
ax = plt.gca()

exportData1.plot(figsize=(15,10), ax=ax, label = 'Keeling')
exportData2.plot(figsize=(15,10), ax=ax, label = 'Law Dome')
exportData3.plot(figsize=(15,10), ax=ax, label = 'Dome C')

ax.set_xlim(-800000,2000)
plt.title('CO2 Historical Record')
plt.show()
```

### Exporting Data

```python
exportData = pd.concat([exportData1, exportData2, exportData3]).sort_index(ascending = False)
exportData['YearsBefore2019'] = 2019 - exportData.index
exportData
```

### Plot Combined Exported Data

```python
exportData.plot(y = 'CO2', figsize=(15,10), xlim=[10000,-800000])
plt.show()
```

### Export Data

```python
exportData.to_csv('CO2.csv')
```
