```python
import pandas as pd
import matplotlib.pyplot as plt
```

# EPICA Dome C Ice Core Temperature Data
### Used for past 800 thousand years 

Data accessed from https://www.ncdc.noaa.gov/paleo-search/study/6080

Note that BP = Before Present = Before 1950. So before exporting the data we need to convert the year to be consistent with the other data sets

```python
domec = pd.read_csv ('Dome C Ice Core 800KYr Temperature Data.txt', delim_whitespace=True)
domec['YearAD'] = 1950 - domec['Age']
domec = domec.set_index('YearAD')
exportData = domec[['Temperature']].iloc[12:]
exportData
```

### Plot Temperature Data

```python
exportData.plot(figsize=(15,10))
plt.title('Historical Temperature Data')
plt.show()
```

### Export Data

```python
exportData['yearsBefore2019'] = 2019 - exportData.index
exportData.columns = ['temp', 'yearsBefore2019']
exportData.index.names = ['year']
exportData
```

```python
exportData.to_csv('Temp.csv')
```
